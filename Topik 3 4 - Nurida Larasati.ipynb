# Basic Data Operation

You sold 20 pieces of electronics at the price of 12000. The customer gets a total discount of 10%.

Below are those variables:

1.   price=12000
2.   quantity_sold=20
3.   discount=10%


Using Python code, write the code the find the total money that the customer to pay. (Points : 15)

```python
# membuat variabel price berisi integer 12000
price = 12000
# membuat variabel quantity_sold berisi integer 20
quantity_sold = 20
# membuat variabel discount berisi integer 0.1
discount = 0.1

# menghitung total sebelum diskon dengan perkalian price dikali quantity_sold
total = price * quantity_sold
# menghitung final_pay dengan total dikali 1 yang dikurangi diskon
final_pay = total * (1-discount)
print( 'the total money that the customer to pay',final_pay)
```

# Dictionary Data Operation

6. >  Follow the instructions below





```python
# Given the dictionary below

customer={
    "name":['Dio','Eka','Wayu','Riki'],
    "age":[20,19,24,39],
    "gender":[1,0,0,1],
    "membership":['Platinum','Gold','Silver','Silver']
}
```

```python
# Print all keys in customer (Points : 5)

customer
```

```python
# Print all customer names (Points : 5)

# mengembalikan value dari key 'name' dengan kode dibawah ini
customer['name']
```

```python
# Print the first customer name (Points : 5)

# menggunakan key 'name' pada index ke 1 untuk mengembalikan nama customer pertama
customer['name'][0]
```

```python
# Drop "age" key from the dictionary (Points : 5)

# menggunakan fungsi pop untuk menghilangkan key 'age' pada dictionary customer
# mengganti value dengan None
customer.pop('age',None)
customer
```

```python
# Replace customer gender list value (Points : 20)
# gender 1 => Male
# gender 0 => Female 


# membuat fungsi untuk mengganti value pada key 'gender'
def change_gender(gender):
    # apabila gender memiliki value 1 maka mengembalikan 'Male'
    if gender == 1:
        return 'Male'
    # apabila gender memiliki value 1 maka mengembalikan 'Female'
    else:
        return 'Female'

# mengganti key 'gender' dengan list baru
# menggunakan fungsi map untuk menggunakan fungsi change_gender yang dibuat sebelumnya
customer['gender'] = list(map(change_gender, customer['gender']))
customer
```
