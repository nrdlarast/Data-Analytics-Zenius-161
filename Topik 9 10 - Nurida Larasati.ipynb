In this assignment, we will use the dataset of diabetes patients taken [from here](https://www4.stat.ncsu.edu/~boos/var.select/diabetes.html).

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')


df = pd.read_csv("https://www4.stat.ncsu.edu/~boos/var.select/diabetes.tab.txt",sep='\t')
df.head()
```


In this dataset, columns as the following:
* Age and sex are self-explanatory
* BMI is body mass index
* BP is average blood pressure
* S1 through S6 are different blood measurements
* Y is the qualitative measure of disease progression over one year

Let's study this dataset using methods of probability and statistics.

### Mean

Mean adalah nilai rata-rata dari suatu kumpulan data. Biasanya digunakan untuk menggambarkan nilai pusat atau nilai tengah dari suatu distribusi data. 

```python
np.mean(df)
```

### Variance

Variance adalah ukuran statistik yang mengukur seberapa jauh nilai-nilai dalam suatu dataset tersebar atau tersebar dari nilai mean. Variansi digunakan untuk menggambarkan seberapa besar variasi atau fluktuasi data dalam distribusi. 

```python
np.var(df)
```

## **Task 2: Plot boxplots for BMI, BP and Y depending on gender**

image_2023-04-18_184317793.png

Boxplot atau Box-Plot merupakan visualisasi data berbentuk box (kotak berbentuk bujur sangkar. Boxplot dapat digunakan untuk menunjukkan perbedaan antara populasi tanpa menggunakan asumsi distribusi statistik yang mendasarinya.

Boxplots dapat membantu pengguna untuk memahami karakteristik dari distribusi data.

### Boxplot BMI

```python
plt.figure(figsize=(5, 6))
sns.boxplot(data=df,
            y = 'BMI',
            x = 'SEX',
            palette="bright",
            orient = 'v')

q1_1 = df[df['SEX'] == 1]['BMI'].quantile(0.25)
q2_1 = df[df['SEX'] == 1]['BMI'].quantile(0.5)
q3_1 = df[df['SEX'] == 1]['BMI'].quantile(0.75)

q1_2 = df[df['SEX'] == 2]['BMI'].quantile(0.25)
q2_2 = df[df['SEX'] == 2]['BMI'].quantile(0.5)
q3_2 = df[df['SEX'] == 2]['BMI'].quantile(0.75)

print('Kuartil BMI untuk SEX sama dengan 1: Q1 = {}, Q2 = {}, Q3 = {}'.format(q1_1,q2_1,q3_1))
print('Kuartil BMI untuk SEX sama dengan 2: Q1 = {}, Q2 = {}, Q3 = {}'.format(q1_2,q2_2,q3_2))
```

### Boxplot BP

```python
plt.figure(figsize=(5, 6))
sns.boxplot(data=df,
            y = 'BP',
            x = 'SEX',
            palette="bright")

q1_1 = df[df['SEX'] == 1]['BP'].quantile(0.25)
q2_1 = df[df['SEX'] == 1]['BP'].quantile(0.5)
q3_1 = df[df['SEX'] == 1]['BP'].quantile(0.75)

q1_2 = df[df['SEX'] == 2]['BP'].quantile(0.25)
q2_2 = df[df['SEX'] == 2]['BP'].quantile(0.5)
q3_2 = df[df['SEX'] == 2]['BP'].quantile(0.75)

print('Kuartil BP untuk SEX sama dengan 1: Q1 = {}, Q2 = {}, Q3 = {}'.format(q1_1,q2_1,q3_1))
print('Kuartil BP untuk SEX sama dengan 2: Q1 = {}, Q2 = {}, Q3 = {}'.format(q1_2,q2_2,q3_2))
```

### Boxplot Y

```python
plt.figure(figsize=(5, 6))
sns.boxplot(data=df,
            y = 'Y',
            x = 'SEX',
            palette="bright")

q1_1 = df[df['SEX'] == 1]['Y'].quantile(0.25)
q2_1 = df[df['SEX'] == 1]['Y'].quantile(0.5)
q3_1 = df[df['SEX'] == 1]['Y'].quantile(0.75)

q1_2 = df[df['SEX'] == 2]['Y'].quantile(0.25)
q2_2 = df[df['SEX'] == 2]['Y'].quantile(0.5)
q3_2 = df[df['SEX'] == 2]['Y'].quantile(0.75)

print('Kuartil Y untuk SEX sama dengan 1: Q1 = {}, Q2 = {}, Q3 = {}'.format(q1_1,q2_1,q3_1))
print('Kuartil Y untuk SEX sama dengan 2: Q1 = {}, Q2 = {}, Q3 = {}'.format(q1_2,q2_2,q3_2))
```

* BMI adalah Indeks massa tubuh
* BP adalah tekanan darah rata-rata
* Y adalah ukuran kualitatif perkembangan penyakit selama satu tahun

```python
sns.set(style="whitegrid") # Set style whitegrid untuk tampilan boxplot
plt.figure(figsize=(10, 6)) # Set ukuran figure

# Menggunakan seaborn boxplot untuk menggabungkan BMI, BP, dan Y, berdasarkan hue Sex
sns.boxplot(x='variable', y='value', hue='SEX', data=df[['BMI', 'BP', 'Y', 'SEX']].melt(id_vars='SEX'), palette='bright')

# Memberikan judul dan label pada axis
plt.title('Boxplot BMI, BP, dan Y berdasarkan Sex')
plt.xlabel('Variabel')
plt.ylabel('Nilai')

# Menampilkan legend
plt.legend(title='Sex')

# Menampilkan plot
plt.show()
```
